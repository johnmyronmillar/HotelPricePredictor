# Linear Regression Model

```python
# Import libraries

import numpy as np
import pandas as pd
import hvplot.pandas
from collections import Counter
from pathlib import Path
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error 

import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the cleaned hotel booking data into the workbook and convert to pandas Dataframe
city_csv_path = Path('../Resources/City_hotel_Prep.csv')
resort_csv_path = Path('../Resources/Resort_Hotel_Prep.csv')

city_df = pd.read_csv(city_csv_path)
resort_df = pd.read_csv(resort_csv_path)
```

# City Data

```python
# Drop unamed row
city_df = city_df.drop(columns = ['Unnamed: 0', 'hotel'])
resort_df = resort_df.drop(columns = ['Unnamed: 0', 'hotel'])
```

```python
city_df.head()
```

```python
city_df.corr()
```

```python
ax = sns.heatmap(city_df.corr())
```

```python
# Create a scatter plot of the data
city_df.hvplot(kind = 'hist', title = 'ADR Instances', x = 'date', y = 'adr', ylabel = 'Instances', xlabel = "ADR Range")
```

```python
# filter city data into adr and week number

city_avg_week = city_df.filter(['adr','arrival_date_week_number'], axis=1)
avg_weekly_adr = city_avg_week.groupby('arrival_date_week_number').mean()
avg_weekly_adr.head()
```

```python
# filter city data into adr and date
city_avg = city_df.filter(['adr','date'], axis=1)

# convert date to datetime and set index to date
city_avg['date']= pd.to_datetime(city_avg['date'])
city_avg = city_avg.set_index('date')

# resample city data (weekly) to determine average price per date
city_avg_price = city_avg.resample('W').mean().dropna()
city_avg_price.head()
```

```python
avg_weekly_adr.head()
```

```python
avg_weekly_adr.hvplot(kind = 'line', title = 'City Weekly Average Price by Week Number', ylabel = 'Avg ADR', xlabel = "Week Number")
```

```python
# Plot the weekly average price
city_avg_price.hvplot(kind = 'line', title = 'City Weekly Average Price by Date', x = 'date', y = 'adr', ylabel = 'Avg ADR', xlabel = "Date")
```

```python
# Plot the weekly average price
city_avg_price.hvplot(kind = 'scatter', title = 'City Weekly Average Price by Date', x = 'date', y = 'adr', ylabel = 'ADR', xlabel = "Date")
```

```python
# Plot lead time and adr
city_df.hvplot(kind = 'scatter', title = 'City Lead Time & ADR', x = 'lead_time', y = 'adr', ylabel = 'ADR', xlabel = "Lead Time (Days)")
```

```python
# filter city data into adr and week number

city_lead_time = city_df.filter(['adr','lead_time'], axis=1)
city_avg_lead = city_lead_time.groupby('lead_time').mean()
city_avg_lead.head()
```

```python
city_avg_lead.hvplot(kind = 'scatter', title = 'Lead Time and Price', ylabel = 'Avg ADR', xlabel = "Lead Time")
```

```python
# Create a Linear Regression Model using sklearn
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model
```

```python
# Separate the data into the X and y components
X = city_avg_lead.index.values.reshape(-1,1)
y = city_avg_lead['adr']
```

```python
# Create training and testing datasets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2)
```

```python
# Fit the model to the data
model.fit(X_train, y_train)
```

```python
# Make predictions using the X data
predicted_y_values_city = model.predict(X_test)
```

```python
plt.scatter(X, y)
plt.plot(X_test, predicted_y_values_city, color='red')
```

```python
# Assemble actual y data (Y_test) with predicted y data (from just above) into two columns in a dataframe:
city_results = pd.DataFrame(y_test)
city_results["Predicted Value"] = predicted_y_values_city
# Results.sort_values(["arival_date"], axis=0, ascending=True, inplace=True)
city_results['Diff'] = city_results['Predicted Value'] - city_results.adr
city_results.head(5)
```

```python
print(city_results.Diff.abs().mean())
print(city_results.Diff.abs().max())
print(city_results.Diff.abs().min())
print(city_results.Diff.abs().std())
```

```python
from sklearn.metrics import mean_squared_error
city_mse = mean_squared_error(y_test, predicted_y_values_city)
city_mse = city_mse ** (1/2)
print(city_mse)
```
